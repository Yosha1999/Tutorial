```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
```

```python
x = np.arange(0, math.pi*2, 0.05)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine')
plt.title('Sine Wave')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.set_title('Sine Wave')
ax.plot(x,y)
```

```python
fig = plt.figure()
x = np.arange(0, math.pi*2, 0.05)
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
ax3 = fig.add_axes([0.17, 0.17, 0.3, 0.3])
ax1.plot(x, np.sin(x), 'b')
ax1.set_title('Sin')
ax2.plot(x, np.cos(x), 'r')
ax2.set_title('Cos')
ax3.plot(x, np.tan(x), 'k')
ax3.set_title('Tan')
plt.show()
```

```python
fig, a = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x, x*x, 'b')
a[0][0].set_title('Square')
a[0][1].plot(x, np.sqrt(x), 'r')
a[0][1].set_title('Square root')
a[1][0].plot(x, np.exp(x), 'c')
a[1][0].set_title('Exponential')
a[1][1].plot(x, np.log10(x), 'm')
a[1][1].set_title('Logarithm')
plt.tight_layout()
```

```python
x = np.arange(1,10)
a1 = plt.subplot2grid((4,4),(0,0), colspan=3)
a2 = plt.subplot2grid((4,4),(0,3), rowspan=4)
a3 = plt.subplot2grid((4,4),(1,0), rowspan= 3,colspan=3)
a1.plot(x, x*x, 'b')
a1.set_title('Square')
a2.plot(x, np.log10(x), 'm')
a2.set_title('Logarithm')
a3.plot(x, np.exp(x), 'c')
a3.set_title('Exponential')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,3, figsize=(12,4))
x = np.arange(1,11)
ax[0].plot(x, x**3, 'g', lw=2)
ax[0].grid()
ax[1].plot(x, np.exp(x), 'r', lw=1)
ax[1].grid(color='b', ls='--', lw=1)
ax[2].plot(x,x)
plt.tight_layout()
```

```python
fig, ax= plt.subplots(1,2, figsize=(10,4))
x = np.arange(1,5)
ax[0].plot(x, x*x)
ax[0].plot(x, np.exp(x))
ax[0].set_xlabel('x-axis')
ax[0].set_ylabel('y-axis')
ax[0].set_title('Normal Scale')
ax[0].xaxis.labelpad = 10
ax[0].yaxis.labelpad = 10
ax[0].spines['top'].set_color('w')

ax[1].plot(x, np.exp(x))
ax[1].plot(x, x*x)
ax[1].set_yscale('log')
ax[1].set_xlabel('x-axis')
ax[1].set_ylabel('y-axis')
ax[1].set_title('Logarithmic Scale')
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.5,0.5])
x = np.arange(1,10)
ax.plot(x, np.exp(x))
ax.set_xlim(0,10)
ax.set_ylim(0,10000)
plt.plot()
```

```python
fig = plt.figure()
x = np.arange(0, math.pi*2, 0.05)
ax = fig.add_axes([0,0,0.5,0.5])
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.set_title('Sine Wave')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero','two','four','six'])
ax.set_yticks([-1,0,1])
ax.plot(x,np.sin(x))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])
x = np.arange(1,11)
ax.plot(x, np.exp(x), 'o-')
ax.set_ylabel('Exp')
a1 = ax.twinx()
a1.plot(x, np.log(x), 'rs--', ms=5)
a1.set_ylabel('Log')
plt.plot()
```
